```r
'''
lets work with data on cats!
'''

str(MASS::cats)
```

```r
hist(MASS::cats$Bwt)

'''
histogram shows the distribution of body weight of cats in our population(144 cats)
'''
```

```r
mean(MASS::cats$Bwt)

'''
this helps us to work out the mean body weight
also known as the population mean, miu
a parameter that is typically unknown
'''
```

```r
# lets sample the body weight of 3 cats at a time
# then we take the average of our sample
# we then store the average as a value in an array 


sample_size <- 30
n <- 1000000
array <- rep(NA, n)

for(i in 1:n){
    array[i] <- mean(sample(MASS::cats$Bwt,sample_size))
}

hist(array)
abline(v=mean(MASS::cats$Bwt))
```

```r
'''
By tweaking the sample size (recall that CLT requires sample size of 30 and above)
we can see that the sample mean distribution resembles that of normal distribution
and the mean of our sample mean approaches miu
'''
```
